```python
# Importing Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datasets import load_dataset

# Loading Data
dataset = load_dataset("husainaliarif/Covid19")
df = dataset["train"].to_pandas()
```

```python
df_GR = df[(df["continent"] == "Europe") & (df["location"] == "Greece")].copy()
df_GR = df_GR.sort_values("date")
df_GR = df_GR.dropna(subset=["new_cases"])

df_GR
```

```python
df_plot = df_GR.head(10)

sns.lineplot(data=df_plot, x="date", y="new_cases")
sns.set_theme(style="ticks")
sns.despine()
plt.xticks(rotation=30)
plt.title("Daily New Covid Cases in Greece")
plt.xlabel("Date")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
cols = ["new_cases", "new_vaccinations", "new_tests"]
df_barplot = df_GR[cols].mean().reset_index()
df_barplot.columns = ["metric", "mean_value"]
df_barplot = df_barplot.sort_values("mean_value", ascending=False)

sns.barplot(data=df_barplot, x="mean_value", y="metric", width=0.5)
sns.set_theme(style="ticks")
sns.despine()
plt.title("Average Daily Covid Metrics in Greece")
plt.xlabel("Mean Value")
plt.ylabel("")
plt.show()
```
